```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
from matplotlib import font_manager, rc
```

```python
data = pd.read_csv('./소고기가격_2016_2022_11월까지.csv')
```

```python
data
```

```python
data['date'] = data['date'].str.replace(' ', '')
data.head()
```

```python
for i in range(len(data['도매가격'])):
    data['도매가격'][i] = data['도매가격'][i].split(' ')[0]
```

```python
data.head()
```

```python
data['도매가격'] = data['도매가격'].replace('-', np.NaN)
data['도매가격'] = data['도매가격'].str.replace(',', '')
data
```

```python
data['date'] = pd.to_datetime(data['date'])
data.info()
```

```python
data['도매가격'] = pd.to_numeric(data['도매가격'])
data.info()
```

```python
start_date='20160104'
end_date='20221117'
date_list=pd.date_range(start=start_date, end=end_date, freq='D')
# print(date_list)
temp = pd.DataFrame(date_list)
temp.rename(columns={0:'date'},inplace=True)
temp
```

```python
data = pd.merge(data,temp,on='date', how='outer').sort_values(by='date')
data
```

```python
data.fillna(method='ffill',inplace=True)
data.isnull().sum()
```

```python
data.to_csv('./소고기가격_최종_2016_2022_11월17일까지.csv')
```
