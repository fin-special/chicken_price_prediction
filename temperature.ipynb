# 월별 기온/닭고기 가격 상관관계 분석

```python
# 2012~2021년 각 월별 평균 기온 데이터
avg_temperature = pd.read_csv('./2012-2021기온.csv', encoding='utf-8')
```

```python
avg_temperature.head()
```

```python
avg_temperature.info()
```

```python
avg_temperature.head()
```

```python
chicken_price = pd.read_csv('/localRepository/recipes_oraganization/chicken_price_prediction/데이터처리/닭고기데이터/닭고기시세20120101_20211231.csv')
```

```python
chicken_price.head()
# 호수별 가격을 일별로 평균 내서, 닭고기 평균 가격을 산정했음
```

```python
chicken_price.rename(columns={'닭고기 평균 가격':'price'}, inplace=True)
```

```python
chicken_price.info()
```

```python
# type을 str에서 datetime으로 바꿈
chicken_price['date'] = pd.to_datetime(chicken_price['date'])
```

```python
# 닭고기 가격을 str에서 int로 바꿈
chicken_price['price'] =  chicken_price['price'].str.replace(',','')
chicken_price['price'] = pd.to_numeric(chicken_price['price'])
```

```python
chicken_price.info()
```

```python
chicken_price
```

```python
chicken_price.set_index(pd.DatetimeIndex(chicken_price['date'])).drop('date',axis=1)
```

```python
df_m = chicken_price.resample('M',on='date').mean()
```

```python
# 월별 평균 계산
df_m
```

```python
df_m.info()
```

```python
# price를 float에서 int로 바꿈
df_m['price'] = df_m['price'].astype('int')
```

```python
df_m
```

```python
df_m = df_m.reset_index()
df_m
```

```python
df_m.to_csv('montly_chicken_price.csv')
```

```python
df = pd.concat([avg_temperature,df_m],axis=1)
df
```

```python
df.drop(['date','최저기온(℃)','최고기온(℃)','지점'],axis=1,inplace=True)
```

```python
df
```

```python
df.corr(method='pearson')
```

# 일별 기온/닭고기 가격 상관관계 분석

```python
temperature_daily = pd.read_csv('./2012-2021 기온 daily.csv')
temperature_daily
```

```python
temperature_daily['날짜'] = temperature_daily['날짜'].str.replace('\t', '')
temperature_daily
```

```python
temperature_daily.isnull().sum()
```

```python
temperature_daily.dropna(axis=0,inplace=True)
temperature_daily
```

```python
temperature_daily.rename(columns={'날짜':'date'},inplace=True)
temperature_daily
```

```python
chicken_price
```

```python
df_daily = pd.merge(temperature_daily,chicken_price,on='date',how='right')
df_daily
```

```python
df_daily = df_daily[['date','평균기온(℃)', 'price']]
df_daily
```

```python
df_daily['price'] = df_daily['price'].str.replace(',','')
df_daily['price'] = df_daily['price'].astype('int')
df_daily.info()
```

```python
df_daily.corr(method='pearson')
```
