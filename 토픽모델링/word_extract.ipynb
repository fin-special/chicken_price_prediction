```python
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import nltk  
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
from konlpy.tag import Twitter, Okt, Hannanum, Kkma, Komoran
from collections import Counter
```

```python
# 뉴스기사 데이터 불러오기
news201201 = pd.read_csv('./data/닭_2012_2021/닭고기_가격_2012년1월뉴스.csv')
news201201.columns
```

```python
filename = '../데이터전처리/data/닭_2012_2021/닭고기_가격_2012년1월뉴스.csv'
f = open(filename, 'r', encoding='utf-8')
news = f.read()
```

```python
# Okt 객체 선언
okt = Okt()

# Kkma 객체 선언
kkma = Kkma()

# Hannanum 객체 선언
hannanum = Hannanum()

# Twitter 객체 선언
twitter = Twitter()

# Komoran 객체 선언
komoran = Komoran()
```

```python
# 명사 추출
words = []

for i in news201201['기사내용']:
    word = kkma.nouns(i)
    for j in word:
        words.append(j)
    
words
```

```python
noun = kkma.nouns(news)
count = Counter(noun)

noun_list = count.most_common(100)
for i in noun_list:
    print(i)
```
