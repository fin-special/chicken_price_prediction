## 정규화
- MinMaxScaler / StandardScaler 두 가지 방법이 있음
- 두 개 다 돌려보고 정확도가 더 높은 함수를 사용함. 보통 잘 모를때는 StandardScaler 사용함. (해보고 별로면 MinMaxScaler 사용)

- 배열화 처리 먼저 해주기 (numpy 사용)

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
# 스케일링 함수
ss = StandardScaler()
```

#### 닭고기 가격 정규화 값 컬럼 추가 (배열화 > 스케일링 > 차원축소 > 컬럼추가)

```python
# 닭고기 가격 데이터 배열화 처리
price_df2 = price_df[['price']].to_numpy()
price_df2
```

```python
ss.fit(price_df2)
priceScale=ss.transform(price_df2)
priceScale
```

```python
# 2차원 -> 1차원 리스트에 담기
price_list=[]
for i in priceScale :
    price_list.append(i[0])

price_list
```

```python
price_df['price_ss'] = price_list
price_df
```

#### 수입량 정규화 값 컬럼 추가 (배열화 > 스케일링 > 차원축소 > 컬럼추가)

```python
# 수입량 데이터 배열화 처리
chick_df2 = chick_df[['병아리(원/수)']].to_numpy()
```

```python
ss.fit(chick_df2)
chickScale=ss.transform(chick_df2)
chickScale
```

```python
# 2차원 -> 1차원 리스트에 담기
chick_list=[]
for i in chickScale :
    chick_list.append(i[0])

chick_list
```

```python
chick_df['chick_ss'] = chick_list
chick_df.head()
```

```python
# 월별 닭고기, 수입량 정규화 그래프

plt.style.use('default')
plt.rcParams['figure.figsize'] = (20,10)
plt.rcParams['font.size'] = 15
plt.rcParams['font.family'] = font_name

x = price_df['date']
y1 = price_df['price_ss']
y2 = chick_df['chick_ss']

fig, ax1 = plt.subplots()
ax1.set_title('월별 닭고기 가격,병아리(원/수) 정규화 그래프')
ax1.set_xlabel('시점')
ax1.set_ylabel('닭고기 가격')
ax1.plot(x, y1, color='blue',label='닭고기')

ax2 = ax1.twinx()
ax2.set_ylabel('병아리(원/수)')
ax2.plot(x, y2, color='red',label='병아리(원/수)')

plt.xticks(['2012-01-31', '2013-01-31', '2014-01-31','2015-01-31','2016-01-31','2017-01-31', '2018-01-31', '2019-01-31', '2020-01-31', '2021-01-31'])
fig.legend()
plt.show()
```
